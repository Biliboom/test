```python
df = data.drop(["Historic", "Classification Key", "Disaster Group", "Disaster Subgroup", "External IDs", "Event Name", "Country", "Region", "Subregion", "Origin", "Associated Types", "OFDA/BHA Response", "Appeal", "Declaration", "AID Contribution ('000 US$)", "Latitude", "Longitude", "River Basin", "No. Homeless", "Reconstruction Costs ('000 US$)", "Reconstruction Costs, Adjusted ('000 US$)", "Insured Damage ('000 US$)", "Insured Damage, Adjusted ('000 US$)", "CPI", "Admin Units"], axis=1)
#Dropping Irrelevant Variables (subject to change)
```

## Example with Brazil

```python
BRA_df= df.loc[(df['ISO'] == "BRA")]       #Brazilian subset of dataframe
```

```python
BRA_df["Disaster Type"].value_counts().plot(kind='bar')     #Brazilian histogram of different disasters
```

```python
BRA_FLD_df = BRA_df.loc[(df['ISO'] == "BRA") & (df["Disaster Type"] == "Flood")]      #Brazilian Floods
```

```python
BRA_FLD_df["Start Year"].hist()       #Histogram of number of floods over time (rough)
```

```python
BRA_STR_df = BRA_df.loc[(df['ISO'] == "BRA") & (df["Disaster Type"] == "Storm")]      #Brazilian Storms
```

```python
BRA_STR_df["Start Year"].hist()    #Histogram of number of storms over time
```
